```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/data.csv')
```

```python
df.head(5)
```

```python
# creating a new column comb , which contains combination of all features
df['comb']=df['actor_1_name']+' '+df['actor_2_name']+' '+df['actor_3_name']+' '+df['director_name']+' '+df['genres']
```

```python
df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count_vector=CountVectorizer()
```

```python
count_matrix=count_vector.fit_transform(df['comb'])
```

```python
count_matrix
```

```python
similar=cosine_similarity(count_matrix)
```

```python
similar
```

```python
# getting the index of the movie in the dataframe
movie_index =  df.loc[df['movie_title']=='avatar'].index[0]
```

```python
# fetching the row containing similarity scores of the movie
lst = list(enumerate(similar[movie_index]))
```

```python
lst
```

```python
lst = sorted(lst, key = lambda x:x[1] ,reverse=True)
```

```python
len(lst[1:11]) #exclude 0 because that is our given movie
```

```python
top_10 = lst[1:11]
```

```python
movies=[]
for i in range(len(top_10)):

    a = top_10[i][0]
    movies.append(df['movie_title'][a])
```

COMBINING ALL DOTS

```python
def top_recomendations(movie):
    movie_index =  df.loc[df['movie_title']==movie].index[0]
    print(movie_index)
    lst = list(enumerate(similar[movie_index]))
    lst = sorted(lst, key = lambda x:x[1] ,reverse=True)
    top_10 = lst[1:11]

    l=[]
    for i in range(len(top_10)):

        a = top_10[i][0]
        l.append(df['movie_title'][a])

    return l

```

```python
print(top_recomendations('spectre'))
```

```python
print(top_recomendations('the dark knight rises'))
```
